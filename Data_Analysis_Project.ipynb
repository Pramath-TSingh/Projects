```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt # visualizing data
%matplotlib inline
import seaborn as sns
```

```python
# import csv file
dataframe = pd.read_csv('Diwali Sales Data.csv', encoding= 'unicode_escape')
```

```python
#for information
dataframe.info()
```

```python
# deleteing the uneeded columns
dataframe.drop(['Status', 'unnamed1'], axis=1, inplace = True)
```

```python
pd.isnull(dataframe)
```

```python
pd.isnull(dataframe).sum()
```

```python
dataframe.dropna(inplace=True)
```

```python
dataframe.shape
```

```python
pd.isnull(dataframe).sum()
```

```python
#changeing data type
dataframe['Amount'] = dataframe['Amount'].astype('int')
```

```python
# Exploratory Data Analysis
```

```python
ax = sns.countplot(x = 'Gender',hue = 'Gender', data = dataframe)


for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_gen = dataframe.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
px = sns.barplot(x='Gender', y='Amount', hue = 'Gender', data=sales_gen)
```

```python
# From the above graph  you can see most of the buyers are females and even the pyrchasing power of females are greater than men
```

```python
# Ages

ax = sns.countplot(data = dataframe, x = 'Age Group', hue = 'Gender')

for bars in ax.containers:
  ax.bar_label(bars)
```

```python
# Total Amount Vs Age Group

sales_age = dataframe.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x= 'Age Group',y='Amount', hue = 'Age Group',data=sales_age)
```

```python
# From the above graphs we can see the most of the buyers are age group to 26-35 are women
```

```python
# State

# total number of orders from top 10 states
sales_state = dataframe.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data= sales_state, x= 'State', y = 'Orders', hue= 'State')
```

```python
# State

# total number of orders from top 10 states
sales_state = dataframe.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data= sales_state, x= 'State', y = 'Amount', hue= 'State')
```

```python
# Marital Status

ax = sns.countplot(data=dataframe, x='Marital_Status', hue='Marital_Status')

sns.set(rc={'figure.figsize':(15,5)})
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
# State

# total number of orders from top 10 states
sales_state = dataframe.groupby(['Marital_Status','Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data= sales_state, x= 'Marital_Status', y = 'Amount', hue= 'Gender')
```

```python
# Most Shopping done is by Married Women with More Purchasing Power
```

```python
# Occupation

sns.set(rc={'figure.figsize': (20,5)})
ax = sns.countplot(data= dataframe, x='Occupation', hue='Occupation')

for bars in ax.containers:
  ax.bar_label(bars)
```

```python
# Occupation

# total number of orders from top 10 states of Occupation
sales_state = dataframe.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data= sales_state, x= 'Occupation', y = 'Amount', hue= 'Occupation')
```

```python
# From the Above Graphs we cam see the ,pst pf the buyers are woking in IT Sector, HealthCare amd Aviation sector
```

```python
# Occupation

sns.set(rc={'figure.figsize': (20,5)})
ax = sns.countplot(data= dataframe, x='Product_Category', hue='Product_Category')

for bars in ax.containers:
  ax.bar_label(bars)
```

```python
# Product Cat

# total number of orders from top 10 states of AMount
sales_state = dataframe.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data= sales_state, x= 'Product_Category', y = 'Amount', hue= 'Product_Category')
```

```python
# From the Above Graphs we cam see the most of the sold products are from Food, Clothing and Electornics Category
```

```python
# Product ID

# total number of orders from top 10 states of AMount
sales_state = dataframe.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data= sales_state, x= 'Product_ID', y = 'Orders', hue= 'Product_ID')
```

```python
# Top 10 Products sold
fig1, ax1 = plt.subplots(figsize=(12,7))
dataframe.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind = 'bar')
```

```python
# Conclusion

# Married women age group from 26-35 years from UP, Maharastra and Karnatka working in IT, Healthcare and Aviation are
# some likely to buy prodcuts from food, Clothing and Eltronics category.
```
